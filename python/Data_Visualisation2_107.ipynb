```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 1, 10, 5, 7])
print(plt.plot(ypoints, linestyle='dashed'))

plt.title('Sample graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

print(plt.show())
```
